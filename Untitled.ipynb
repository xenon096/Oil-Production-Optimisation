```python
%matplotlib inline
pyplot.style.use('ggplot')
def emperical(data):
    n=len(data)
    x=numpy.sort(data)
    y=numpy.arange(1,n+1)/n
    return x,y
#plotting w1
x_ax,y_ax=emperical(file_w1['BORE_OIL_VOL'])
pyplot.plot(x_ax,y_ax,marker=".",linestyle='none')
#label
pyplot.xlabel('Water Injection(Sm3)')
pyplot.ylabel('Distribution')
pyplot.title("Bore:5693(w1)")
pyplot.show()
```
