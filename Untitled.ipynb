```python
import numpy
from matplotlib import pyplot
%matplotlib inline
pyplot.style.use('ggplot')
#Emperical Cumulative Distribution Function
def emperical(data):
    n=len(data)
    x=numpy.sort(data)
    y=numpy.arange(1,n+1)/n
    return x,y

#---------------------------------------------------
#w7
x,y=emperical(file_w7['BORE_OIL_VOL'])
pyplot.plot(x,y,marker='.',linestyle='none')
#labeling
pyplot.title('Bore no:7405(w7)')
pyplot.xlabel('Volume(Sm3)')
pyplot.ylabel('Distribution')

pyplot.show()
```

```python
#round off values of data for integer
file=numpy.round(file)
```

```python
#as the given data is skewed witn null values-- padding them(Forward Filling)
file['ON_STREAM_HRS'] = file['ON_STREAM_HRS'].fillna(method='pad')

file['AVG_DOWNHOLE_PRESSURE'] = file['AVG_DOWNHOLE_PRESSURE'].fillna(method='pad')

file['AVG_DOWNHOLE_TEMPERATURE'] = file['AVG_DOWNHOLE_TEMPERATURE'].fillna(method='pad')

file['AVG_DP_TUBING'] = file['AVG_DP_TUBING'].fillna(method='pad')

file['AVG_ANNULUS_PRESS'] = file['AVG_ANNULUS_PRESS'].fillna(method='pad')

file['AVG_CHOKE_SIZE_P'] = file['AVG_CHOKE_SIZE_P'].fillna(method='pad')

file['AVG_WHP_P'] = file['AVG_WHP_P'].fillna(method='pad')

file['AVG_WHT_P'] = file['AVG_WHT_P'].fillna(method='pad')

file['DP_CHOKE_SIZE'] = file['DP_CHOKE_SIZE'].fillna(method='pad')

file['BORE_OIL_VOL'] = file['BORE_OIL_VOL'].fillna(method='pad')

file['BORE_GAS_VOL'] = file['BORE_GAS_VOL'].fillna(method='pad')

file['BORE_WAT_VOL'] = file['BORE_WAT_VOL'].fillna(method='pad')
```

```python
#Removing difference in distribution in columns to scale dataset
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
file[['ON_STREAM_HRS','AVG_DOWNHOLE_TEMPERATURE','AVG_ANNULUS_PRESS','AVG_CHOKE_SIZE_P','AVG_WHP_P','AVG_WHT_P']] = scaler.fit_transform(file[['ON_STREAM_HRS', 'AVG_DOWNHOLE_TEMPERATURE','AVG_ANNULUS_PRESS', 'AVG_CHOKE_SIZE_P','AVG_WHP_P', 'AVG_WHT_P']])
#Taking the data of wells:5693 & 5769 since they were the ones with more number of missing values
ff=file.loc[746:8022]

x=ff.drop(['BORE_OIL_VOL'],axis=1)
y= ff[['BORE_OIL_VOL','NPD_WELL_BORE_CODE']]
#Applying Linear Regression
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)

L_reg=LinearRegression()
```

```python
#gaining date values of daily production
print(file.DATEPRD)
```

```python
#now taking the dates of production for training the data on daily basis for x variable
final_x_train=x_train.drop(['DATEPRD','NPD_WELL_BORE_CODE'],axis = 1)
final_x_test=x_test.drop(['DATEPRD',"NPD_WELL_BORE_CODE"],axis = 1)
```

```python
#taking the dates of production for training the data on daily basis for y variable
final_y_train=y_train['BORE_OIL_VOL']
final_y_test=y_test['BORE_OIL_VOL']
```

```python
#gaining x,y variables of the data collected on wells 5693,5769
x=ff.drop(['DATEPRD','BORE_OIL_VOL'],axis=1).values
y=ff['BORE_OIL_VOL']
```

```python
#getting final x trained value
final_x_train
```

```python
#regression of x,y trained data 
L_reg.fit(final_x_train,final_y_train)
```

```python
#prediction of y variable with R2 linear regression trained x test value
pred_y=L_reg.predict(final_x_test)
```

```python
print('R2 linear regression for oil volume production:',L_reg.score(final_x_test,final_y_test))
```

```python
#getting count with enumerate to get summary of the given regression with coef_
stat=L_reg.coef_

for k,v in enumerate(stat):
    print('No:%d, score:%.5f'%(k,v))
```
